```python
import tkinter
import tkinter.font
import random, time

COLORS = ['red', 'blue', 'green', 'pink', 'orange', 'purple']
TABOOS = ['hat','this','Robert','under','backpack','granola','tire',
          'dick','cunt','nigger','table','fart','heck','dingus','dickhole',
          'ass','God','book','friend','into','sheared','shit','poop','retard',
          'cancer','putain']*2

words_in_order = []
times = []
correct = []
last_color = ''

def stimulus():
    global COLORS, TABOOS, words_in_order,last_color
    try:
        word = random.choice(TABOOS)
        words_in_order.append(word)
        TABOOS.remove(word)
    except IndexError:
        return('done','black')
    color = random.choice(COLORS)
    while color == last_color:
        color = random.choice(COLORS)
    last_color = color
    return (word, color)

def next_selected(color):
    global times, correct
    times.append(time.time())
    if label.cget("fg") == color:
        correct.append(1)
    else:
        correct.append(0)
    word, color = stimulus()
    label.config(text=word, fg=color)
    label.update()
    if label.cget("text") == 'done':
        for element in root.winfo_children():
            root.unbind('r')
            root.unbind('b')
            root.unbind('g')
            root.unbind('p')
            root.unbind('v')
            root.unbind('o')
            element.destroy()
        quit_selected()
        return

def quit_selected():
    """gives you the stats, times are in milliseconds"""
    last = times[0]
    newt = []
    for time in times[1:]:
       newt.append((time-last)*1000)
       last = time
    tuples  = zip(words_in_order,newt,correct)
    tp  = ["|     Word     | Time (ms)  | Correctness |"]
    for item in tuples:
        tp.append(''.join(['| ',item[0]," "*(13-len(item[0])),'| ',
            str(round(item[1],0))," "*(11-len(str(round(item[1],0)))),'| ',
            str(item[2])," "*(12-len(str(item[2]))),'|']))
    #results = tkinter.Label(root,text='\n'.join(tp),fg='black',bg='white',justify='left',font=my_font)
    results = tkinter.Listbox(root,font=my_font)
    for item in tp:
        results.insert('end',item)
    results.config(width=0)
    scrollbar = tkinter.Scrollbar(root)
    scrollbar.pack(side='right', fill='y')
    scrollbar.config(command=results.yview)
    results.config(yscrollcommand=scrollbar.set) 
    results.pack()

root = tkinter.Tk()
root.configure(background='white')
my_font = tkinter.font.Font(family='Consolas',size=12)


# create the window

# create label using stimulus
word, color = stimulus()

label = tkinter.Label(root, text=word, fg=color, bg='white',font=my_font)
label.grid(row = 0, column = 1, padx = 20, pady=20)

red = tkinter.Button(root,text = 'red',command= lambda: next_selected('red'),font=my_font)
blue = tkinter.Button(root,text = 'blue',command= lambda: next_selected('blue'),font=my_font)
green = tkinter.Button(root,text = 'green',command= lambda: next_selected('green'),font=my_font)
black = tkinter.Button(root,text = 'pink',command= lambda: next_selected('pink'),font=my_font)
orange = tkinter.Button(root,text = 'orange',command= lambda: next_selected('orange'),font=my_font)
purple = tkinter.Button(root,text = 'violet',command= lambda: next_selected('purple'),font=my_font)

root.bind('r',lambda event: next_selected('red'))
root.bind('b',lambda event: next_selected('blue'))
root.bind('g',lambda event: next_selected('green'))
root.bind('p',lambda event: next_selected('pink'))
root.bind('o',lambda event: next_selected('orange'))
root.bind('v',lambda event: next_selected('violet'))


red.grid(row = 1, column = 0, padx = 20, pady=20)
blue.grid(row = 1, column = 1, padx = 20, pady=20)
green.grid(row = 1, column = 2, padx = 20, pady=20)
black.grid(row = 2, column = 0, padx = 20, pady=20)
orange.grid(row = 2, column = 1, padx = 20, pady=20)
purple.grid(row = 2, column = 2, padx = 20, pady=20)

times.append(time.time())
root.mainloop()

```
